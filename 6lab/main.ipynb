```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
terms_n = [5, 7]
for n in terms_n:
    f = open(f'ga_error_for_terms/{n}_terms.txt', 'r')
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i] = float(lines[i])
    plt.plot(lines)
    plt.xlabel('generation')
    plt.ylabel('error')
    plt.title(f'GA: error by generetion for {n} terms')
    plt.show()
```

```python
population_sizes = [10, 20, 30]
for n in population_sizes:
    f = open(f'ga_error_for_generations/{n}_gens.txt', 'r')
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i] = float(lines[i])
    r = np.linspace(0, 30, 4)
    ys = []
    xs = []
    for i in range(len(r)-1):
        y = lines[int(r[i]):int(r[i+1])]
        x = [r[i+1]] * len(y)
        xs.extend(x)
        ys.extend(y)
    plt.scatter(xs, ys, label=f'population_size={n}')
plt.xlabel('generation')
plt.ylabel('error')
plt.legend()
plt.title(f'GA: distribution of errors by generation for 5 terms for several populations sizes')
plt.show()
```
