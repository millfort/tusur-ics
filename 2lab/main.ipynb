```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
from skfuzzy import control as ctrl
from rich import print as pprint
from IPython.display import display, Markdown
from tabulate import tabulate
```

```python
class Bounds:
    def __init__(self, left, right):
        self.left = left
        self.right = right

X_BOUNDS = Bounds(-5., 5.)
Y_BOUNDS = Bounds(-5., 5.)
Z_BOUNDS = Bounds(-1., 1.)

PI = 3.1415926535897932384626433832795

def F(x, y):
    return np.sin(2*x/PI) * np.sin(2*y/PI)
```

```python
x = np.linspace(X_BOUNDS.left, X_BOUNDS.right, 50)
y = np.linspace(Y_BOUNDS.left, Y_BOUNDS.right, 50)
X, Y = np.meshgrid(x, y)

def to_bound(low, high):
    def f(v):
        if v < low:
            return low
        if v > high:
            return high
        return v
    return f

def make_z(x, y, f, lb, rb):
    z = f(x, y)
    for i, zz in enumerate(z):
        zz = list(map(to_bound(lb, rb), zz))
        z[i] = zz
    return z

Z = make_z(X, Y, F, Z_BOUNDS.left, Z_BOUNDS.right)

def plot_3d_surface(x, y, z):  
    ax = plt.axes(projection='3d')
    plt.title(' : sin(2x/pi)*sin(2y/pi)')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.plot_wireframe(x, y, z)

plot_3d_surface(X, Y, Z)
```

```python

terms = [
    [
        ' ',
        '',
        '',
        '',
        ' '
    ],
    [
        ' ',
        ' ',
        '',
        '',
        '',
        ' ',
        ' '   
    ],
    [
        ' ',
        ' ',
        ' ',
        '',
        '',
        '',
        ' ',
        ' ',
        ' ', 
    ],
    [
        ' ',
        ' ',
        ' ',
        '',
        ' ',
        '',
        ' ',
        '',
        ' ',
        ' ',
        ' ', 
    ]
]
```

```python
for ts in terms:
    display(Markdown(f'# Result for {len(ts)} terms'))
    step = 0.05
    x_universe = np.arange(X_BOUNDS.left, X_BOUNDS.right+step, step)
    y_universe = np.arange(Y_BOUNDS.left, Y_BOUNDS.right+step, step)
    z_universe = np.arange(Z_BOUNDS.left, Z_BOUNDS.right+step, step)

    xv = ctrl.Antecedent(x_universe, 'x')
    yv = ctrl.Antecedent(y_universe, 'y')
    zv = ctrl.Consequent(z_universe, 'z')

    xv.automf(len(ts), names=ts)
    yv.automf(len(ts), names=ts)
    zv.automf(len(ts), names=ts)

    display(Markdown(f'## Member functions'))

    fuzz.control.visualization.FuzzyVariableVisualizer(xv).view()
    fuzz.control.visualization.FuzzyVariableVisualizer(yv).view()
    fuzz.control.visualization.FuzzyVariableVisualizer(zv).view()
    plt.show()

    def gen_abcs(universe, number, bl, br):
        limits = [universe.min(), universe.max()]
        universe_range = limits[1] - limits[0]
        widths = [universe_range / ((number - 1) / 2.)] * int(number)
        centers = np.linspace(limits[0], limits[1], number)
        abcs = [[c - w / 2, c, c + w / 2] for c, w in zip(centers, widths)]
        abcs[0][0] = bl
        abcs[-1][-1] = br
        return abcs

    x_params = gen_abcs(x_universe, len(ts), X_BOUNDS.left, X_BOUNDS.right)
    y_params = gen_abcs(y_universe, len(ts), Y_BOUNDS.left, Y_BOUNDS.right)
    z_params = gen_abcs(z_universe, len(ts), Z_BOUNDS.left, Z_BOUNDS.right) 

    for name, params in [["x", x_params], ["y", x_params], ["z", z_params]]:
        display(Markdown(f'## Parameters of the membership functions for {name}'))
        print(tabulate(params, floatfmt=".2f", headers=['1', '2', '3']))

    def find_term(universe, antecedent, v):
        max_deg = 0
        max_term = ''
        for term in antecedent.terms:
            deg = fuzz.fuzzymath.interp_membership(
                universe, 
                antecedent[term].mf, 
                v
            )
            if deg <= max_deg:
                continue
            max_deg = deg
            max_term = term
        return max_term

    vals = {}
    to_z_bound = to_bound(Z_BOUNDS.left, Z_BOUNDS.right)
    for xi in x_universe:
        x_term = find_term(x_universe, xv, xi)
        for yi in y_universe:
            y_term = find_term(y_universe, yv, yi)
            z_vals = vals.get((x_term, y_term))
            if z_vals == None:
                z_vals = {}
                vals[(x_term, y_term)] = z_vals
            zi = F(xi, yi)
            zi = to_z_bound(zi)
            z_term = find_term(z_universe, zv, zi) 
            z_vals[z_term] = z_vals.get(z_term, 0) + 1

    def to_singleton(terms, term):
        return terms.index(term) - ((len(terms) - 1) / 2)

    def calc_z(xtria, ytria):
        x = (xtria[0] + xtria[1])/2
        y = (ytria[0] + ytria[1])/2
        return F(x, y)
    # display(Markdown(f'## Rules'))

    # for k in vals:
    #     zterm = max(vals[k], key=vals[k].get)
    #     print(f'IF x IS {k[0]:15} AND y IS {k[1]:15} THEN z IS {zterm:15} (F = {to_singleton(list(zv.terms), zterm):4})')

    display(Markdown(f'## Rule base '))
    too = []
    for i, x_param in enumerate(x_params):
        for j, y_param in enumerate(y_params):
            too.append([i+1, j+1, calc_z(x_param, y_param)])
    print(tabulate(too, headers=['x', 'y', 'z']))
    
```
